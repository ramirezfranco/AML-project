# Topic modeling with out city names

```python
cities_ = [t for v in cities.values() for t in nltk.word_tokenize(v['name'])]
states_ = [t for v in list(states_corpus.keys()) for t in nltk.word_tokenize(v)]
places = list(set(cities_+states_))
```

```python
states_corpus_bis = {}
for k, v in states_corpus.items():
    tokens = v.split()
    valid_words=[t for t in tokens if t not in places]
    states_corpus_bis[k] = ' '.join(valid_words)
```

```python
states_topics_bis = tm.Topics(states_corpus_bis, 15)
```

```python
states_topics_bis.z_matrix('lda')
states_topics_bis.z_matrix('nmf')
states_topics_bis.z_matrix('lsi')
print()
```

```python
for i,r in states_topics_bis.topic_words('lda').iterrows():
    print(i, '|', r['Content'])
    print('**********************************************************')
```

```python
for i,r in states_topics_bis.topic_words('nmf').iterrows():
    print(i, '|', r['Content'])
    print('**********************************************************')
```
