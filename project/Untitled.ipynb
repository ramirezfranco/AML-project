```python
cities = util.open_json('data/cities.json')
```

```python
states = {}
for city in cities.values():
    if city['state'] not in states.keys():
        states[city['state']]= {'cities':[city['name']]}
    else:
        states[city['state']]['cities'].append(city['name'])
```

```python
#get text of reviews
for v in states.values():
    v['text']= ''
    for city in v['cities']:
        revs = util.get_city_reviews(city)
        v['text']+= ' '+' '.join(list(revs.values()))
```
